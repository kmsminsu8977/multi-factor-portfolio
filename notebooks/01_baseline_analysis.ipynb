```python
portfolio = tables["portfolio_weights"]
display(portfolio.head(10))

summary = pd.Series({
    "selected_count": len(portfolio),
    "weight_sum": portfolio["selected_weight"].sum(),
    "min_weight": portfolio["selected_weight"].min(),
    "max_weight": portfolio["selected_weight"].max(),
    "risk_contribution_sum": portfolio["risk_contribution_pct"].sum(),
    "annualized_volatility": portfolio["portfolio_annualized_volatility"].iloc[0],
})
summary
```
